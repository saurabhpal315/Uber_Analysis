```python
##month column
df['Month']=pd.DatetimeIndex(df['START_DATE']).month
```

```python
df
```

```python
dic = {1: 'jan', 2: 'feb', 3: 'mar', 4: 'apr', 5: 'may', 6: 'jun', 7: 'jul', 8: 'aug', 9: 'sep', 10: 'oct', 11: 'nov', 12: 'dec'}
```

```python
df['Month']=df['Month'].map(dic)
```

```python
df
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x=df['Month'],order=df['Month'].value_counts().index)
```

```python
def round(x):
    if x['START']==x['STOP']:
        return 'yes'
    return 'no'
```

```python
df['Round_trips']=df.apply(round,axis=1)
```

```python
df
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=df['Round_trips'],order=df['Round_trips'].value_counts().index)
```

```python
df['Round_trips'].value_counts()
```

```python
###which month is having more round trips
plt.figure(figsize=(12,7))
sns.countplot(x=df['Round_trips'],hue=df['Month'],palette='viridis')
plt.legend()
```

```python
a=df.groupby(['Month','Round_trips']).agg({'Round_trips':'count'})
```
